```python
import torch
import psutil
from diffusers import AutoPipelineForText2Image

# Function to get available system memory
def get_available_system_memory():
    memory_info = psutil.virtual_memory()
    return memory_info.available / (1024 ** 3)  # Convert to GB

# Function to get available GPU memory
def get_available_gpu_memory():
    if torch.cuda.is_available():
        torch.cuda.synchronize()
        gpu_memory_allocated = torch.cuda.memory_allocated()
        total_gpu_memory = torch.cuda.get_device_properties(0).total_memory
        return (total_gpu_memory - gpu_memory_allocated) / (1024 ** 3)  # Convert to GB
    return 0

# Set up model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
pipeline = AutoPipelineForText2Image.from_pretrained("black-forest-labs/FLUX.1-dev", torch_dtype=torch.float16)

# Enable model offloading
pipeline.enable_model_cpu_offload()
pipeline.enable_sequential_cpu_offload()

# Optional: Enable slicing for memory efficiency during VAE operations
pipeline.vae.enable_slicing()
pipeline.vae.enable_tiling()

# Check memory usage before training
available_cpu_memory = get_available_system_memory()
available_gpu_memory = get_available_gpu_memory()

print(f"Available system RAM: {available_cpu_memory:.2f} GB")
print(f"Available GPU RAM: {available_gpu_memory:.2f} GB")

# You can define a memory threshold and decide when to shift load between CPU and GPU
SYSTEM_RAM_THRESHOLD_GB = 10  # Example threshold for CPU memory
GPU_RAM_THRESHOLD_GB = 5  # Example threshold for GPU memory

# Example: Shifting workload based on memory availability
if available_gpu_memory < GPU_RAM_THRESHOLD_GB:
    print("Not enough GPU memory, offloading more to CPU")
    pipeline.to('cpu')
else:
    print("Running model on GPU")
    pipeline.to(device)


```

```python
import torch

def get_gpu_memory():
    if torch.cuda.is_available():
        # Get total and allocated GPU memory
        total_memory = torch.cuda.get_device_properties(0).total_memory
        allocated_memory = torch.cuda.memory_allocated()
        free_memory = total_memory - allocated_memory

        return {
            'total_memory': total_memory / (1024 ** 3),  # Convert to GB
            'allocated_memory': allocated_memory / (1024 ** 3),  # Convert to GB
            'free_memory': free_memory / (1024 ** 3),  # Convert to GB
        }
    else:
        return "CUDA is not available."

memory_info = get_gpu_memory()
print(f"Total GPU Memory: {memory_info['total_memory']:.2f} GB")
print(f"Allocated GPU Memory: {memory_info['allocated_memory']:.2f} GB")
print(f"Free GPU Memory: {memory_info['free_memory']:.2f} GB")
```

```python
import os

# Folder containing the text files
prompt_folder = '/content/drive/MyDrive/Create_Dataset/Create_Dataset/textfile_folder'
# Folder to save the modified text files
output_folder = '/content/drive/MyDrive/Create_Dataset/Create_Dataset/textfilekeyword_folder'

# Make sure the output folder exists; if not, create it
os.makedirs(output_folder, exist_ok=True)

# Words to append
words_to_append = 'pattern, textile, designs, motif'

# Iterate through all files in the prompt_folder
for filename in os.listdir(prompt_folder):
    # Only process .txt files
    if filename.endswith('.txt'):
        file_path = os.path.join(prompt_folder, filename)

        # Read the original file content
        with open(file_path, 'r') as file:
            content = file.read()

        # Append the words to the file content
        updated_content = content.strip() + ' ' + words_to_append

        # Save the updated content to the output folder with the same filename
        output_file_path = os.path.join(output_folder, filename)
        with open(output_file_path, 'w') as output_file:
            output_file.write(updated_content)

print("Text files have been processed and saved with updated content.")
```
